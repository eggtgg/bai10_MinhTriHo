```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
name_friend=["Trí","Thịnh","Huy","Hoàng","Nhật","Minh"]
cash=[1,2,3,4,2,3]
card=[3,4,2,1,4,1]
```

```python
list_name=np.arange(len(name_friend))
```

```python
plt.xticks(list_name,name_friend)
plt.ylabel("(Tỷ Đồng)")
plt.title('Số Tiền hiện có')
plt.bar(list_name-0.2,cash, width=0.4, label= 'Trong Ví')
plt.bar(list_name+0.2,card, width=0.4, label= 'Trong thẻ')
plt.legend()
```

```python
plt.yticks(list_name,name_friend)
plt.ylabel("(Tỷ Đồng)")
plt.title('Số Tiền hiện có')
plt.barh(list_name-0.2,card, label= 'Trong Ví')
plt.barh(list_name+0.2,cash, label= 'Trong thẻ')
plt.legend()
```
